```python
from math import pi 
radius = eval(input("Enter the radius value ="))

if radius > 0 :
    area = pi * radius * radius
    print("Area of circle is =" , format(area, ".2f"))
    circumference = 2*pi*radius
    print("circumference of the cirle =", format(circumference, ".2f"))
else:
    print("please give the right input")
    
```
